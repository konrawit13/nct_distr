```python
df.loc[df['ProductTradeName'].str.contains('alprazolam', case=False)]
```

```python
df_agg = df.loc[df['ProductTradeName'].str.contains('alprazolam', case=False)].groupby(by=["CustomerNameTh"])['มูลค่า'].sum().reset_index().sort_values(by=['มูลค่า'])
```

```python
df_agg['CustomerNameTh'].value_counts().reset_index().sort_values(by="count")
```

```python
df.loc[df['ProductTradeName'].str.contains('alprazolam', case=False)]['CustomerNameTh'].value_counts().reset_index()
```

```python
df_cn_ct = df.loc[df['ProductTradeName'].str.contains('alprazolam', case=False)]['CustomerNameTh'].value_counts()
```

```python
df_pamount_sum = df.loc[df['ProductTradeName'].str.contains('alprazolam', case=False)].groupby(by=["CustomerNameTh"])['ProductAmount'].sum().reset_index()
```

```python
df_pamount_sum
```

```python
df_merge = df_agg.merge(
    df_cn_ct,
    how ='left',
    on = ['CustomerNameTh']
)
```

```python
df_merge = df_merge.merge(
    df_pamount_sum,
    how ='left',
    on = ['CustomerNameTh']
)
```

```python
df_merge
```

```python
df_alpra = df.loc[df['ProductTradeName'].str.contains('alprazolam', case=False)]
```

```python
df_alpra.columns
```

```python
df_merge2 = df_merge.merge(
    df_alpra.loc[:,['CustomerTypeName', 'CustomerNameTh']].drop_duplicates(),
    how = 'left',
    on = ['CustomerNameTh']
)
```
